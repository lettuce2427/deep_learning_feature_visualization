```python
from scipy.io import loadmat
import numpy as np
import random
def load_data(file_path):
    data = loadmat(file_path)
    data_x = data['X']
    data_y = data['y']
    data_y = np.where(data_y==10, [0],data_y)
    num_data = data_x.shape[3]
    return_data = []
    for i in range(num_data):
        return_data.append(data_x[:,:,:,i])
    return_label = to_categorical(data_y)
    return np.array(return_data), return_label
def to_categorical(y, num_classes=None):
    """
    把输入的标签变成one-hot形式
    """
    y = np.array(y, dtype='int').ravel()
    if not num_classes:
        num_classes = np.max(y) + 1
    n = y.shape[0]
    categorical = np.zeros((n, num_classes))
    categorical[np.arange(n), y] = 1
    return categorical
def get_batch(data, label, batch_size):
    list = [i for i in range(label.shape[0])]
    random.shuffle(list)
    return_data = data[list[:batch_size],:,:,:]
    return_label = label[list[:batch_size],:]
    return return_data, return_label
```

```python
import tensorflow as tf
from tensorflow.contrib import slim
import numpy as np
class CNN():
    def __init__(self, input_size, input_depth, num_classes, learning_rate, is_training=True):
        graph = tf.Graph()
        with graph.as_default():
            self.x = tf.placeholder(tf.float32, shape=[None, input_size, input_size, input_depth])
            self.y_ = tf.placeholder(tf.float32, shape=[None, num_classes])
            self.keep_prob = tf.placeholder(tf.float32)
            self.conv1 = slim.conv2d(self.x, 32, [5, 5], scope='convolution_1')
            self.pool1 = slim.max_pool2d(self.conv1, [2, 2], 2, scope='pool1')
            self.conv2 = slim.conv2d(self.pool1, 64, [5, 5], scope='convolution_2')
            self.pool2 = slim.max_pool2d(self.conv2, [2, 2], 2, scope='pool2')
            flatten = slim.flatten(self.pool2)
            self.fc3 = slim.fully_connected(flatten, 1024, scope='fc3')
            drop_out = slim.dropout(self.fc3, self.keep_prob, is_training=is_training, scope='dropout3')
            self.y = slim.fully_connected(drop_out, num_classes, activation_fn=None, scope='fc4')
            self.loss = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(logits=self.y, labels=self.y_), name='loss')
            self.train_op = tf.train.AdamOptimizer(learning_rate=learning_rate).minimize(self.loss)
            self.predicted = tf.argmax(self.y, 1)
            self.accuracy = tf.reduce_mean(tf.cast(tf.equal(self.predicted, tf.argmax(self.y_, 1)), tf.float32), name='acc')
            init = tf.global_variables_initializer()
            self.saver = tf.train.Saver()
        self.sess = tf.Session(graph=graph)
        self.sess.run([init])

    # conv层的显示
    # 格式是 （？，size, size, depth)
    def show_conv(self, out, fig):
        show_row = out.shape[0]
        show_col = out.shape[3]
        for row in range(show_row):
            for col in range(show_col):
                ax = fig.add_subplot(show_row, show_col, (row * show_col + col + 1))
                show_img = out[row, :, :, col]
                ax.imshow(show_img, )
                ax.axis('off')
    # fc 层显示函数
    # 数据格式是：（？， size)
    def show_fc(self, out, fig):
        show_row = out.shape[0]
        show_size = int(np.sqrt(out.shape[1]))
        for row in range(show_row):
            ax = fig.add_subplot(show_row, 1, row + 1)
            show_img = out[row, :]
            show_img = np.reshape(show_img, [show_size, show_size])
            ax.imshow(show_img)
            ax.axis('off')
```

```python
import tensorflow as tf
from tensorflow.contrib import slim
import numpy as np
class CNN():
    def __init__(self, input_size, input_depth, num_classes, learning_rate):
        self.input_size = input_size
        graph = tf.Graph()
        with graph.as_default():
            self.x = tf.placeholder(tf.float32, shape=[None, input_size, input_size, input_depth])
            self.y_ = tf.placeholder(tf.float32, shape=[None, num_classes])
            self.keep_prob = tf.placeholder(tf.float32)
            # 第一层卷积层
            self.W_conv1 = self.__weight_variable([5, 5, input_depth, 32])
            self.b_conv1 = self.__bias_variable([32])
            self.h_conv1 = tf.nn.relu(self.__conv2d(self.x, self.W_conv1) + self.b_conv1)
            self.h_pool1 = self.__max_pool_2x2(self.h_conv1)

            # 第二层卷积层
            self.W_conv2 = self.__weight_variable([5, 5, 32, 64])
            self.b_conv2 = self.__bias_variable([64])
            self.h_conv2 = tf.nn.relu(self.__conv2d(self.h_pool1, self.W_conv2) + self.b_conv2)
            self.h_pool2 = self.__max_pool_2x2(self.h_conv2)

            flatten = slim.flatten(self.h_pool2)
            self.fc3 = slim.fully_connected(flatten, 1024, scope='fc3')
            drop_out = slim.dropout(self.fc3, self.keep_prob, is_training=True, scope='dropout3')
            self.y = slim.fully_connected(drop_out, num_classes, activation_fn=None, scope='fc4')
            self.loss = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(logits=self.y, labels=self.y_),
                                       name='loss')
            self.train_op = tf.train.AdamOptimizer(learning_rate=learning_rate).minimize(self.loss)
            self.predicted = tf.argmax(self.y, 1)
            self.accuracy = tf.reduce_mean(tf.cast(tf.equal(self.predicted, tf.argmax(self.y_, 1)), tf.float32),
                                           name='acc')
            init = tf.global_variables_initializer()
            self.saver = tf.train.Saver()
        self.sess = tf.Session(graph=graph)
        self.sess.run([init])

    def __weight_variable(self, shape):
        initial = tf.truncated_normal(shape, stddev=0.1)
        return tf.Variable(initial)

    def __bias_variable(self, shape):
        initial = tf.constant(0.1, shape=shape)
        return tf.Variable(initial)

    # 定义卷积操作
    def __conv2d(self, x, W):
        return tf.nn.conv2d(x, W, strides=[1, 1, 1, 1], padding='SAME')

    # 定义池化操作
    def __max_pool_2x2(self, x):
        return tf.nn.max_pool(x, ksize=[1, 2, 2, 1], strides=[1, 2, 2, 1], padding='SAME')
def show_conv1(fig, input_conv, input_kernel, input_size, depth):
    if depth==1:
        cmap_input = 'gray_r'
        cmap_conv = 'gray_r'
        shape = [input_size, input_size]
    if depth == 3:
        cmap_input = None
        cmap_conv = 'gray_r'
        shape = [input_size, input_size, depth]
    sess = tf.Session()
    show_conv_shape = [input_size, input_size]
    convs = [input_conv[:,:,:,i].reshape([1, input_size, input_size, 1]) for i in range(32)]
    kernels =[input_kernel[:,:,:,i].reshape([input_kernel.shape[0], input_kernel.shape[1], input_kernel.shape[2], 1]) for i in range(32)]
    deconv = []
    for i in range(32):
        deconv.append(tf.nn.conv2d_transpose(convs[i], filter=kernels[i], output_shape=[1, input_size, input_size, depth], strides=[1, 1, 1, 1], padding='SAME'))
    deconvs = sess.run([deconv])[0]
    deconv_count = 0
    conv_count = 0
    for row in range(8):
        for col in range(8):
            ax = fig.add_subplot(8, 8, row * 8 + col + 1)
            if col >= 4:
                ax.imshow(norm(deconvs[deconv_count].reshape(shape)), cmap=cmap_input)
                deconv_count = deconv_count + 1
            else:
                ax.imshow(convs[conv_count].reshape(show_conv_shape), cmap=cmap_conv)
                conv_count = conv_count + 1
            ax.axis('off')

def show_conv2(fig, input_conv, input_kernel1, input_kernel2, input_size, depth):
    if depth==1:
        cmap_input = 'gray_r'
        cmap_conv = 'gray_r'
        shape = [input_size, input_size]
    if depth == 3:
        cmap_input = None
        cmap_conv = 'gray_r'
        shape = [input_size, input_size, depth]
    conv_shape = [input_conv.shape[1], input_conv.shape[2]]
    sess = tf.Session()
    conv_2_shape = [1, input_conv.shape[1], input_conv.shape[2], 1]
    conv_2_kernel_shape = [input_kernel2.shape[0], input_kernel2.shape[1], input_kernel2.shape[2], 1]
    deconv_2_shape = [1, input_size, input_size, input_kernel2.shape[2]]

    convs = [input_conv[:, :, :, i].reshape(conv_2_shape) for i in range(64)]
    kernels_2 = [input_kernel2[:, :, :, i].reshape(conv_2_kernel_shape) for i in range(64)]

    deconv_2_to_1 = []
    for i in range(64):
        deconv_2_to_1.append(tf.nn.conv2d_transpose(convs[i], filter=kernels_2[i], output_shape=deconv_2_shape,
                                             strides=[1, 2, 2, 1], padding='SAME'))
    deconv =[tf.nn.conv2d_transpose(deconv_2_to_1[i],filter=input_kernel1, output_shape=[1, input_size, input_size, depth],strides=[1, 1, 1, 1], padding='SAME') for i in range(len(deconv_2_to_1))]
    deconvs = sess.run([deconv])[0]
    print(len(deconv))
    deconv_count = 0
    conv_count = 0
    for row in range(8):
        for col in range(16):
            ax = fig.add_subplot(8, 16, row * 16 + col + 1)
            if col >= 8:
                ax.imshow(norm(deconvs[deconv_count].reshape(shape)), cmap=cmap_input)
                deconv_count = deconv_count + 1
            else:
                ax.imshow(convs[conv_count].reshape(conv_shape), cmap=cmap_conv)
                conv_count = conv_count + 1
            ax.axis('off')
def norm(input_data):
    return (input_data-np.min(input_data))/(np.max(input_data)-np.min(input_data))
```

```python
import numpy as np
from tensorflow.examples.tutorials.mnist import input_data
from cnn2 import CNN

mnist = input_data.read_data_sets("MNIST_data/", one_hot=True)

# 这里各种参数的设置
# input_size 图像的尺寸
# depth 是图像的通道数
# nu_classes 是数据类别
# learning_rate 是学习率
# model_dir 保存模型的地址
input_size = 28
depth = 1
num_classes = 10
learning_rate = 1e-4
model_dir = './train/model.ckpt'
max_step = 500
#
cnn = CNN(input_size, depth, num_classes, learning_rate)

for i in range(max_step):
    # 输入的训练数据  需要转化数据的格式
    batch = mnist.train.next_batch(50)
    batch_x, train_y = batch
    train_x=np.reshape(batch_x,[-1,input_size,input_size,depth])
    #训练
    _, train_loss, train_acc = cnn.sess.run([cnn.train_op, cnn.loss, cnn.accuracy], feed_dict={cnn.x:train_x, cnn.y_:train_y, cnn.keep_prob:0.5})
    print('step:{}, loss:{:0.3f}, acc:{:0.3f}'.format(i, train_loss, train_acc))
cnn.saver.save(cnn.sess, model_dir)
test_images = np.reshape(mnist.test.images, [-1, input_size, input_size, depth])
test_labels = mnist.test.labels
test_acc = cnn.sess.run([cnn.accuracy], feed_dict={cnn.x:test_images, cnn.y_:test_labels, cnn.keep_prob:1.0})
print('test acc:{}'.format(test_acc))

```
